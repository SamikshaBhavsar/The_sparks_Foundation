```
data['State'].value_counts().mean()

```

**City wise analysis of Profits and sale:**



```
data['City'].value_counts()
```

```
City = data.groupby("City")[["Sales", "Profit"]].sum().sort_values(by = "Sales", ascending = False)
City[:30].plot(kind = "bar", color = ["green", "red"],figsize= (18,8))
plt.title("Profit, Loss & Sales of top 30 Cities")
plt.xlabel("Cities")
plt.ylabel("Total profit / loss and sales")
plt.show()
```

***New york has maximum sales and profit***

```
City[-30:].plot(kind = "bar", color = ["green", "red"],figsize= (18,8))
plt.title("Profit, Loss & Sales of economic weak Cities")
plt.xlabel("Cities")
plt.ylabel("Total profit / loss and sales")
plt.show()
```

***Cities like Romeoville , Deer Park, Tinley Park are facing too much loss.***

**State wise analysis of Profit and Sales:**

```
data.groupby('State')[['Profit', 'Sales']].sum().plot.bar(color = ["red", "green"],figsize= (18,8))
plt.ylabel("Total profit / loss and Sales")
plt.show()
```

***California state has maximum sales with good profit whereas many states such as Texas and Ohio face loss.***





**Category wise analysis of Profit and Sales:**

```
data.groupby('Category')[['Profit', 'Sales']].sum().plot.bar(color = ["brown", "pink"],figsize= (8,5))
plt.ylabel("Total profit / loss and Sales")
plt.show()
```

**Sub-Category wise analysis of Profit and Sales:**

```
data.groupby('Sub-Category')[['Profit', 'Sales']].sum().plot.bar(color = ["blue", "orange"],figsize= (12,5))
plt.ylabel("Total profit / loss and Sales")
plt.show()
```

***From the above plot we can conclude that products such as Bookcases and Tables are facing looses and Company have a lot of profit in the sale of Copiers but the numbers of sales is very less so we need to increase the number of sales of copier.***

**Segment wise analysis of profit and sales:**

```
data.groupby('Segment')[['Profit', 'Sales']].sum().plot.bar(color = ["blue", "red"],figsize= (8,5))
plt.ylabel("Total profit / loss and Sales")
plt.show()
```

***We can conclude that sales and profit is maximum for Consumer and minimum for Home Office***

**Region wise analysis of profit and sales:**


```
data.groupby('Region')[['Profit', 'Sales']].sum().plot.bar(color = ["orange", "green"],figsize= (8,5))
plt.ylabel("Total profit / loss and Sales")
plt.show()
```

***Western regions show maximum sales and profit. Wheres Southern regions show minimum sales with relative good profit.***

## **Insights derived from above EDA :**

1. Western regions show maximum sales and profit. Wheres Southern regions show minimum sales with relative good profit. So, for better business sales should be increased in southern regions.
2. Maximum number of deals are from California and New York. We should expand in other parts of USA as well.
3. Furniture supplies have low profit relative to its sale. Profit should be maximised in this category.
4. Improvements should be made for same day shipment mode.
5. Company is facing losses in sales of bookcases and tables products.
6. Company have much profit in the sale of copier but the number of sales is very less so there is a need of increase in number of sales of copier.
7. Profit and Sales are related to each other but they are not directly proportional.Hence we have to work on both sales and profit seperately to increase them.
8. Profit and discount show very weak and negative relationship. This should be kept in mind that before taking any other decision related to business.
9. States like Ohio and Texas are facing high losses, improvement should be made in such states.
